```python
import matplotlib.pyplot as plt
import numpy as np
import pds
```

```python
n = np.arange(-6, 12)
x1 = pds.delta(n)
plt.stem(n, x1);
```

```python
x2 = pds.u(n)
plt.stem(n, x2);
```

```python
x3 = lambda n: np.exp(-.25 * n) * (pds.u(n)-pds.u(n-6))
```

```python
plt.stem(n, x3(n));
```

```python
plt.stem(n, x3(-n));
```

```python
n = np.arange(31)
x4 = np.sin(0.5*n)
plt.stem(n, x4);
```
