```python
import matplotlib.pyplot as plt
import numpy as np
import pds

N = 1000
n = np.arange(N)
n1= np.arange(N+1)

ws = np.array(np.pi*np.arange(1, 6)/6)

x = np.zeros(N)
for w in ws:
    x += np.cos(w*n)
    
        
NN = 3*N
X=pds.fft(x, NN) [:NN//2]
plt.stem(np.abs(x)/N);
```

```python
A = np.array([2,0,1/2,1,5])
A = np.append(A,0)

wk = (ws[:-1] + ws[1:])/2
wk = np.append(wk, np.pi)

M=100
m=np.arange(M+1)

h5 = 5*(pds.sinc(n1-(M/2), wk[4])-pds.sinc(n1-M/2, wk[3]))
h4 = 1*(pds.sinc(n1-(M/2), wk[3])-pds.sinc(n1-M/2, wk[2]))
h3 = 0.5*(pds.sinc(n1-(M/2), wk[2])-pds.sinc(n1-M/2, wk[1]))
h2 = 0*(pds.sinc(n1-(M/2), wk[1])-pds.sinc(n1-M/2, wk[0]))
h1 = 2*(pds.sinc(n1-(M/2), wk[0]))
h = h1+h2+h3+h4+h5


y = pds.lfilter(h, 1, x)
Y = pds.fft(y, 3*N) [:NN//2]
plt.stem(np.abs(Y)/N);
```
