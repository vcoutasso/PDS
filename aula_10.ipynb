```python
import matplotlib.pyplot as plt
from scipy.fft import fft
import numpy as np
import pds
```

```python
Fs = 1e3
step = 0.001
t = np.linspace(0, 1-step, num=int(1/0.001))
x = np.cos(2*np.pi*100*t)+np.sin(2*np.pi*202.5*t);
```

```python
xdft = fft(x)
xdft = xdft[:int(x.size/2)+1]
xdft = xdft/x.size
xdft[1:-1] *= 2
freq = np.linspace(0, Fs/2, num=int(Fs/2)+1)

plt.xlabel('Hz')
plt.ylabel('Amplitude')
plt.plot(freq, abs(xdft))
plt.plot(freq, np.ones(int(len(x)/2)+1));
```

```python
lpad = 2 * x.size
xdft = fft(x, lpad)
xdft = xdft[:int(lpad/2)+1]
xdft = xdft / x.size
xdft[1:-1] *= 2
freq = np.linspace(0, Fs/2, num=int(Fs+1))

plt.xlabel('Hz')
plt.ylabel('Amplitude')
plt.plot(freq, abs(xdft))
plt.plot(freq, np.ones(int(len(x))+1));
```
