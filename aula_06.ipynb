```python
import matplotlib.pyplot as plt
import numpy as np
import pds

w = np.linspace(-np.pi, np.pi, 1024)

L = 11

A = np.sin(w*L/2)/np.sin(w/2)

psi = -w*(L-1)/2

X = A*np.exp(1j*psi)
```

```python
plt.plot(w, np.abs(X));
```

```python
plt.plot(w, np.angle(X));
```

```python
plt.plot(w, A);
```

```python
plt.plot(w, psi);
```

```python
wc = np.pi/4
n = np.arange(-20, 21)
x = pds.sinc(n, wc)

plt.stem(n, x);
```

## DTFT

```python
L = 11
n = np.arange(-10, 21)
# Retangulo usando funcao degrau
x = pds.u(n) - pds.u(n-L)

X, w = pds.dtft(x, n)

plt.plot(w, X);
```

```python
Xa = np.exp(-1j*w*(L-1)/2) * np.sin(w*L/2)/(w/2)

pds.mse(X, Xa)
```

```python
x = pds.u(n)

X, w = pds.dtft(x, n)

plt.plot(w, X);
```
