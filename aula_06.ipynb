```python
import matplotlib.pyplot as plt
import numpy as np
import pds

w = np.linspace(-np.pi, np.pi, 1024)

L = 11

A = np.sin(w*L/2)/np.sin(w/2)

psi = -w*(L-1)/2

X = A*np.exp(1j*psi)
```

```python
plt.plot(w, np.abs(X));
```

```python
plt.plot(w, np.angle(X));
```

```python
plt.plot(w, A);
```

```python
plt.plot(w, psi);
```

## DTFT

```python
import warnings
warnings.filterwarnings("ignore")

L = 11
n = np.arange(-10, 21)
# Retangulo usando funcao degrau
x = pds.u(n) - pds.u(n-L)

X, w = pds.dtft(x, n)

plt.plot(w, np.abs(X));
plt.plot(w, np.angle(X));

Xa = np.exp(-1j*w*(L-1)/2) * np.sin(w*L/2)/np.sin(w/2)
mse = pds.mse(X, Xa)
print("MSE =", mse)
```

```python
x = pds.delta(n)

X, w = pds.dtft(x, n)

plt.plot(w, X);

Xa = 1
mse = pds.mse(X, Xa)
print("MSE =", mse)
```

```python
n0 = 10
x = pds.delta(n - n0)

X, w = pds.dtft(x, n)

plt.plot(w, np.abs(X));
plt.plot(w, np.angle(X));

Xa = np.exp(-1j*w*n0)
mse = pds.mse(X, Xa)
print("MSE =", mse)
```
