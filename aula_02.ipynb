```python
import matplotlib.pyplot as plt
import numpy as np
import pds
```

```python
h = np.array([1,1,1,1])

n= np.arange(-3, 20)
x= pds.delta(n)
y= pds.conv(x, h)

x2 = pds.u(n)
y2 = pds.conv(x2, h)
```

```python
plt.stem(n, x);
```

```python
plt.stem(n, y);
```

```python
plt.stem(n, x2);
```

```python
plt.stem(n, y2);
```

```python
a = np.array([1., .75])
b = np.array([1.])

y3 = pds.lfilter(b, a, x)

plt.stem(n, y3);
```

```python
y3 = pds.slow_filter(b, a, x)

plt.stem(n, y3);
```
